## 项目1

根据生态分布，生成两组考试成绩，并计算相关统计量：平均分、最高分、最低分、标准分、标准差等。



```python
import numpy as np

# 创建符合正太分布的数
mu, sigma = 0.0, 1.0
g = np.random.normal(loc=mu, scale=sigma, size=1000)
g1 = np.random.randn(1000)

# 创建学科成绩
g = (g - g.min()) / 10
phy = np.round(((100.1 - 20) * g + 20), 1)
g1 = (g1 - g1.min()) / 10
mat = np.round(((100.1 - 20) * g1 + 20), 1)
```

Numpy中常用的统计量



```python
# 计算phy的平均值
np.mean(phy)
```

```python
# 标准差
np.std(phy)
```

```python
# 最值
np.min(phy), np.max(phy)
```

```python
# 80%分位数
np.percentile(phy, 80)
```

```python
# 中位数
np.median(phy)
```

```python
# 将两个数组合并，称为二维数组
marks = np.vstack((phy, mat))
marks
```

```python
#沿1轴方向计算平均值，得到phy和mat分别的平均值
np.mean(marks, axis=1)     
```

```python
 #沿0轴方向计算平均值，得到每个人两个学科的平均值
np.mean(marks, axis=0)   
```

```python
np.std(marks, axis=1)    #沿1轴方向的标准差
```

```python
# phy中高于平均值的
phy[phy > np.mean(phy)]
```

```python
# 等效于条件语句的三元操作
np.where(phy > mat, 1, -1)
```

```python
# 返回符合条件的元素
np.where(phy > phy.mean())
```

```python
# 多分支判断和选择
# 对成绩按照规则替换为等级
# 平均分以上20分为A，平均分以上为B
grand_a = np.mean(phy) + 20
grand_b = np.mean(phy)
np.select([phy >= grand_a, 
           np.logical_and(phy >= grand_b, phy < grand_a)], 
          ['A', 'B'], default='C')
```

## 项目5

统计结果和发现规律

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

filename = "/Users/qiwsir/Documents/Codes/DataAnalysis/chapter05/snell.csv"     #数据地址
datas = pd.read_csv(filename)
datas.head()
```

```python
# 利用线性回归，寻找上述两个特征之间的关系

# 划分训练集和测试集
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(datas['alpha'], datas['belta'], 
                                                    test_size=0.4, random_state=40)
```

```python
# 绘图，观察数据分布规律
plt.scatter(X_train, Y_train)
```

```python
import statsmodels.api as sm    #常用的统计模块

X = pd.DataFrame({"c": np.ones(60).T, "x": X_train})    #假设模型为 y=cx^0 + ax^1,将X作为系数
model = sm.OLS(Y_train, X)     #最小二乘法
result = model.fit()    #训练模型
result.params     #得到了各项的系数
```

```python
# y = 0.076 + 0.472x，根据上述结果得到的多项式
# 绘图，观察效果
coe = result.params

fig, ax = plt.subplots()

# 原数据散点图
ax.scatter(X_train, Y_train)

# 多项式图线
x = np.linspace(0, 1.8, 100)
y = coe['c'] + coe['x']*x   
ax.plot(x, y, color='red')
```

```python
# 改进：二次函数

XX_train = pd.DataFrame({"c": np.ones(60).T, "x":X_train, "x**2": X_train*X_train})
result2 = sm.OLS(Y_train, XX_train).fit()
result2.params
```

```python
# 二次式的模型效果
coe = result2.params

fig, ax = plt.subplots()

ax.scatter(X_train, Y_train)

x = np.linspace(0, 1.8, 100)
y = coe['c'] + coe['x']*x + coe['x**2']*x*x
ax.plot(x, y, color='red')
```

```python
# 利用测试集检验模型

# 模型1
coe = result.params
y = coe['c'] + coe['x']*X_test
np.corrcoef(y, Y_test)    #计算相关系数
```

```python
# 模型2
coe = result2.params
y = coe['c'] + coe['x']*X_test + coe['x**2']*X_test*X_test
np.corrcoef(y, Y_test)
```

```python
result.summary()
```

```python
result2.summary()    # R-squared:判定系数，也称为拟合度。越大，模型拟合的越好。
```

```python
# 真实值和预测值之间的差，就叫做残差（英语：residual）
# 利用残差进行检验
# 绘制残差图
import seaborn as sns
sns.residplot(x=X_train.values, y=result2.resid.values)
```

## 项目6

使用scikit-learn实现线性回归

```python
# 制造线性数据集，以便学习使用
err = np.random.RandomState(1)
x = 10 * err.rand(200)
y = 2 * x - 5 + err.randn(200)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=40)
```

```python
from sklearn.linear_model import LinearRegression    #引入模块

model = LinearRegression(fit_intercept=True)    #创建模型
model.fit(x_train[:, np.newaxis], y_train)    #用训练集训练模型
```

```python
print("slope: ", model.coef_[0])    # 变量系数
print("intercept: ", model.intercept_)    # 截距，常数项
```

```python
# 检验
y_fit = model.predict(x_test[:, np.newaxis])     #①
np.corrcoef(y_fit, y_test)
```

```python
# 残差图
resid = y_test - y_fit
sns.residplot(x=x_test, y=resid)
```
